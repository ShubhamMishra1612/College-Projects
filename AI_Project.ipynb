# 1. Bellman Ford Algorithm

Objective :- To find shortest path from one node to rest of the nodes in the graph

Advantage :- Can handle negative weight cycles unlike Dijkstra Algorithm

```python
infinity = float("inf")
```

```python
graph1 = {

        'S': [(8, 'E'), (10, 'A')],
        'A': [(2, 'C')],
        'B': [(1, 'A')],
        'C': [(3, 'B')],
        'D': [(4, 'A'), (1, 'C')],
        'E': [(1, 'D')]

 }
```

```python
graph2 ={
        'S': [(8, 'E'), (10, 'A')],
        'A': [(-2, 'C')],
        'B': [(1, 'A')],
        'C': [(-2, 'B')],
        'D': [(-4, 'A'), (-1, 'C')],
        'E': [(1, 'D')],
}
```

```python
def Bellman_Ford(graph,start):
  shortest_paths = {}

  for nodes in graph:
    shortest_paths[nodes] = infinity

  shortest_paths[start] = 0
  size = len(graph)

  for _ in range(size - 1):
        for node in graph:
            for edge in graph[node]:
                cost = edge[0]
                to_node = edge[1]
                if shortest_paths[node] + cost < shortest_paths[to_node]:
                    shortest_paths[to_node] = shortest_paths[node] + cost


  for node in graph:
        for edge in graph[node]:
            cost = edge[0]
            to_node = edge[1]
            if shortest_paths[node] + cost < shortest_paths[to_node]:
                return 'INVALID - negative cycle detected'

  return shortest_paths












```

# 2. Dijkstra algorithm

Dijkstra Algorithm follows greedy approach to find shortest path from one node to rest of the nodes in a given graph

```python
def dijkstra(G, start='S'):
    shortest_paths = {}
    unvisited = list(G.keys())

    for node in unvisited:
        shortest_paths[node] = infinity

    shortest_paths[start] = 0

    while unvisited:
        min_node = None

        for node in unvisited:
            if min_node is None:
                min_node = node
            elif shortest_paths[node] < shortest_paths[min_node]:
                min_node = node

        for edge in G[min_node]:
            cost = edge[0]
            to_node = edge[1]

            if cost + shortest_paths[min_node] < shortest_paths[to_node]:
                shortest_paths[to_node] = cost + shortest_paths[min_node]

        unvisited.remove(min_node)

    return shortest_paths
```

# 3. Ant Colony Algorithm

```python
import numpy as np
import random

def gen_edges(graph):
    edges = []
    for node in graph:
        for neighbor in graph[node]:
            edges.append((node, neighbor))
    return edges

def create_pheromone_matrix(graph, initial_pheromone=1.0):
    pheromones = {}
    for edge in gen_edges(graph):
        pheromones[edge] = initial_pheromone
    return pheromones

def update_pheromones(pheromones, paths, decay_factor=0.1):
    for edge in pheromones:
        pheromones[edge] *= (1.0 - decay_factor)
    for path in paths:
        path_length = len(path) - 1
        for i in range(path_length):
            edge = (path[i], path[i+1])
            pheromones[edge] += 1.0 / path_length

def find_path(graph,start,end,path = []):
  path = path + [start]
  if(start==end):
    return [path]
  paths = []
  for node in graph[start]:
    if(node not in path):
      new_p = find_path(graph,node,end,path)
      for new in new_p:
        paths.append(new)
  return paths

def choose_path(graph, pheromones, start, end, alpha=1.0, beta=1.0):
    possible_paths = find_path(graph, start, end)
    probabilities = [calculate_probability(pheromones, path, alpha, beta) for path in possible_paths]
    chosen_path = random.choices(possible_paths, probabilities)[0]
    return chosen_path

def calculate_probability(pheromones, path, alpha, beta):
    probability = 1.0
    path_length = len(path) - 1
    for i in range(path_length):
        edge = (path[i], path[i+1])
        probability *= pheromones[edge] ** alpha
    return probability

def ant_colony_optimization(graph, start, end, iterations=100, ants=10, alpha=1.0, beta=1.0, decay_factor=0.1):
    pheromones = create_pheromone_matrix(graph)
    shortest_path = None
    shortest_distance = float('inf')

    for iteration in range(iterations):
        ant_paths = [choose_path(graph, pheromones, start, end, alpha, beta) for _ in range(ants)]

        if not ant_paths:
            continue

        update_pheromones(pheromones, ant_paths, decay_factor)

        # Find the best path among ant paths
        current_shortest_path = min(ant_paths, key=lambda path: calculate_path_length(graph, path))
        current_shortest_distance = calculate_path_length(graph, current_shortest_path)

        if current_shortest_distance < shortest_distance:
            shortest_path = current_shortest_path
            shortest_distance = current_shortest_distance

    return shortest_path, shortest_distance

def calculate_path_length(graph, path):
    length = 0
    for i in range(len(path) - 1):
        length += 1
    return length

graph = {
    "a": ["b", "c", "d"],
    "b": ["a", "c", "h"],
    "c": ["a", "b", "d", "h", "e"],
    "d": ["a", "c", "e", "f"],
    "e": ["c", "d", "f", "g", "h"],
    "f": ["d", "e", "g"],
    "g": ["h", "e", "f"],
    "h": ["c", "g", "e", "b"],
}

start_node = "a"
end_node = "g"

shortest_path, shortest_distance = ant_colony_optimization(graph, start_node, end_node)
print("Shortest Path:", shortest_path)
print("Shortest Distance:", shortest_distance)
```

# 4. Topological Sorting

```python
from collections import defaultdict

# Class to represent a graph
class Graph:
	def __init__(self, vertices):
		self.graph = defaultdict(list) # dictionary containing adjacency List
		self.V = vertices # No. of vertices

	# function to add an edge to graph
	def addEdge(self, u, v):
		self.graph[u].append(v)

	# A recursive function used by topologicalSort
	def topologicalSortUtil(self, v, visited, stack):

		# Mark the current node as visited.
		visited[v] = True

		# Recur for all the vertices adjacent to this vertex
		for i in self.graph[v]:
			if visited[i] == False:
				self.topologicalSortUtil(i, visited, stack)

		# Push current vertex to stack which stores result
		stack.append(v)

	# The function to do Topological Sort. It uses recursive
	# topologicalSortUtil()
	def topologicalSort(self):
		# Mark all the vertices as not visited
		visited = [False]*self.V
		stack = []

		# Call the recursive helper function to store Topological
		# Sort starting from all vertices one by one
		for i in range(self.V):
			if visited[i] == False:
				self.topologicalSortUtil(i, visited, stack)

		# Print contents of the stack
		print(stack[::-1]) # return list in reverse order

# Driver Code
if __name__ == '__main__':
	g = Graph(6)
	g.addEdge(5, 2)
	g.addEdge(5, 0)
	g.addEdge(4, 0)
	g.addEdge(4, 1)
	g.addEdge(2, 3)
	g.addEdge(3, 1)

	print("Following is a Topological Sort of the given graph")

	# Function Call
	g.topologicalSort()


```
